```python
def get_value_counts(df, column):
    return df[column].value_counts()

def annotate_percentages(ax, column):
    total = len(column)
    for p in ax.patches:
        percentage = f'{100 * p.get_height() / total:.1f}%'
        x = p.get_x() + p.get_width() / 2
        y = p.get_height()
        ax.annotate(percentage, (x, y), ha='center', va='bottom')
```

# MACHINE LEARNING

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, LabelEncoder

from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier, StackingClassifier, RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
```

```python
columns = [
    'Sex', 'Nation', 'Job', 'Freelance_Status', 'Genetic_Background', 'Therapy', 'Indoor_Days', 
    'Increasing_Stress', 'Habit_Changes', 'Psychological_History', 'Adaptation_Challenges', 'Job_Engagement', 
    'Social_Vulnerability', 'Psych_Evaluation', 'Support_Options'
]

df = pd.get_dummies(train, columns=columns, drop_first=True)
```

```python
test_df = pd.get_dummies(test, columns=columns, drop_first=True)
```

```python
label_mapping = {'High': 2, 'Medium': 1, 'Low': 0}

df['Emotional_Fluctuations'] = df['Emotional_Fluctuations'].map(label_mapping)
```

```python
df['Emotional_Fluctuations']
```

```python
X = df.drop(columns=['Emotional_Fluctuations'])
y = df['Emotional_Fluctuations']
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=2)
```

```python
print(f"Shape X-train: {X_train.shape}")
print(f"Shape X-test : {X_test.shape}")
```

```python
from sklearn.metrics import accuracy_score

def pipeline_processing_classification(model, X_train, X_test, y_train, y_test, *args, **kwargs):
    """
    A function that creates and fits a machine learning pipeline for classification,
    makes predictions, and prints evaluation metrics.
    """

    pipe = Pipeline([('model', model)])
    
    pipe.fit(X_train, y_train)
    
    y_pred = pipe.predict(X_test)

    print(f"Accuracy Score: {accuracy_score(y_test, y_pred)}")
```

## Random Forest

```python
def get_best_parameters(X_train, y_train):
    """
    Grid search to find the best hyperparameters for a Random Forest Regressor
    """
    random_forest = RandomForestRegressor()

    rf_pipe = Pipeline([
        ('model', random_forest)
    ])
    
    params = {
        'model__n_estimators': [20, 50, 120],
        'model__random_state': [0, 42],
        'model__max_depth': [10, 15, None],
        'model__max_samples': [0.75, 1.0],
        'model__max_features': [0.6, 1.0],
    }
    
    rf_grid = GridSearchCV(
        estimator=rf_pipe,
        param_grid=params,
        cv=5,
        verbose=2,
        n_jobs=-1
    )

    rf_grid.fit(X_train, y_train)
    best_score = rf_grid.best_score_
    best_params = rf_grid.best_params_

    return best_score, best_params
```

```python
# get_best_parameters(X_train, y_train)
```

```python
random_forest = RandomForestClassifier(
    max_depth=15,
    max_features=0.6,
    n_estimators=160,
    random_state=42,
)
```

```python
result = pipeline_processing_classification(random_forest, X_train, X_test, y_train, y_test)
```
